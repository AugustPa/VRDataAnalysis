```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np

# File paths
file_paths = [
    '/Users/apaula/Nextcloud/locustVR/locustVR_data/20231113_150126/20231113_150126_Optomotor_VR1_.csv',
    '/Users/apaula/Nextcloud/locustVR/locustVR_data/20231113_150126/20231113_150126_Optomotor_VR2_.csv',
    '/Users/apaula/Nextcloud/locustVR/locustVR_data/20231113_150126/20231113_150126_Optomotor_VR3_.csv',
    '/Users/apaula/Nextcloud/locustVR/locustVR_data/20231113_150126/20231113_150126_Optomotor_VR4_.csv'
]

# Initialize a Plotly figure
fig = go.Figure()

# Loop through each file, load the data, and add a trace to the figure
for i, file_path in enumerate(file_paths):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Assuming the structure is similar to df_gratings and the relevant column is 'GameObjectRotY'
    time_index = np.array(range(len(df)))
    
    # Add trace
    fig.add_trace(go.Scatter(x=time_index, y=df['GameObjectRotY'], mode='markers', name=f'VR{i+1}'))

# Update layout for a cleaner look
fig.update_layout(
    title='Grating Rotation Over Time Across Different VR Sets',
    xaxis_title='Time Step',
    yaxis_title='Rotation (degrees)',
    legend_title='VR Sets',
    template='plotly_dark' # Feel free to change the template
)

# Show the interactive plot
fig.show()
```
