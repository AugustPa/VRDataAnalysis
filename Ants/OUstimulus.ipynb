```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace with your actual path or just drag-and-drop the file in a Python environment
file_path = "/Users/apaula/Library/Application Support/InsectVRLab/EIXAMPLE/Ou_15018010310.txt"

df = pd.read_csv(file_path)
```

```python
# Force columns to numeric (in case they're objects/strings)
df["Time"] = pd.to_numeric(df["Time"], errors="coerce")
df["CurrentBlue"] = pd.to_numeric(df["CurrentBlue"], errors="coerce")

# Drop any rows that failed conversion (NaN times)
df = df.dropna(subset=["Time", "CurrentBlue"])
```

```python
df = df.sort_values(by="Time")
```

```python
plt.plot(df["Time"], df["CurrentBlue"])
plt.xlabel("Time (sec)")
plt.ylabel("Blue")
plt.ylim(90, 210)
plt.title("right cylinder OU Process shifting to blue")
plt.show()
```
